| Rating | Meaning | Target Audience / Description |
|:--------|:---------|:------------------------------|
| **TV-MA** | Mature Audience | Adults only (17+). Contains strong language, violence, or mature themes. |
| **TV-14** | Parents Strongly Cautioned | Suitable for ages 14+. Parental guidance advised. |
| **TV-PG** | Parental Guidance | Some material may not be suitable for children. |
| **TV-Y7** | Directed to Older Children | Suitable for children age 7 and older. |
| **TV-Y7-FV** | Fantasy Violence | For ages 7+, includes mild fantasy/action violence. |
| **TV-Y** | All Children | Appropriate for all ages. |
| **TV-G** | General Audience | Suitable for all audiences; no offensive content. |
| **G** | General Audiences | Suitable for everyone. |
| **PG** | Parental Guidance | Some content may not be suitable for young children. |
| **PG-13** | Parents Strongly Cautioned | Ages 13+. May include brief strong language or mature themes. |
| **R** | Restricted | Ages 17+. Contains adult material (strong language, violence, etc.). |
| **NC-17** | Adults Only | Ages 18+. Explicit content not suitable for minors. |
| **NR** | Not Rated | Not officially rated by a board. |
| **UR** | Unrated | No rating was assigned or reviewed. |


| New Category          | Included Ratings                             | Meaning                          |
| :-------------------- | :------------------------------------------- | :------------------------------- |
| **Kids**              | TV-Y, TV-Y7, TV-Y7-FV, TV-G                  | Safe for all children            |
| **Family / General**  | G, PG, TV-PG                                 | Suitable for family viewing      |
| **Teens**             | PG-13, TV-14                                 | Suitable for teens (13–16)       |
| **Adults**            | R, NC-17, TV-MA                              | Adult content (17+)              |
| **Unrated / Unknown** | NR, UR, or incorrect values (e.g., "74 min") | No official rating or data error |


```python
def simplify_genre(genre_str):
    if pd.isna(genre_str):
        return 'Unknown'
    
    genre_str = genre_str.lower()
    
    if 'documentar' in genre_str or 'docuseries' in genre_str:
        return 'Documentary'
    elif 'comedi' in genre_str or 'stand-up' in genre_str:
        return 'Comedy'
    elif 'drama' in genre_str:
        return 'Drama'
    elif 'romantic' in genre_str or 'romance' in genre_str:
        return 'Romance'
    elif 'action' in genre_str or 'adventure' in genre_str:
        return 'Action'
    elif 'horror' in genre_str or 'thriller' in genre_str:
        return 'Horror/Thriller'
    elif 'crime' in genre_str:
        return 'Crime'
    elif 'kids' in genre_str or 'children' in genre_str or 'family' in genre_str:
        return 'Kids & Family'
    elif 'reality' in genre_str:
        return 'Reality TV'
    elif 'sci-fi' in genre_str or 'fantasy' in genre_str:
        return 'Sci-Fi & Fantasy'
    elif 'music' in genre_str or 'musical' in genre_str:
        return 'Music'
    elif 'sport' in genre_str:
        return 'Sports'
    elif 'faith' in genre_str or 'spiritual' in genre_str:
        return 'Faith & Spiritual'
    elif 'classic' in genre_str or 'cult' in genre_str:
        return 'Classic/Cult'
    elif 'anime' in genre_str:
        return 'Anime'
    elif 'international' in genre_str or 'foreign' in genre_str:
        return 'International'
    elif 'lgbtq' in genre_str:
        return 'LGBTQ'
    else:
        return 'Other'
```
